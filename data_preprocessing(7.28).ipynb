```python
clutch_time = pd.read_excel("1.xlsx")
```

```python
def prepocessing(excel):
    excel.columns = excel.iloc[1,:]
    excel = excel.drop([0,1])
    return excel
```

```python
b = prepocessing(clutch_time)
for i in range(2,10):
    a = pd.read_excel("%i.xlsx"%i)
    b = pd.merge(b,prepocessing(a),on=["Player","Season","Season Type","Team","Games"],how="left")
b.to_csv("nba_all.csv")
b
```
